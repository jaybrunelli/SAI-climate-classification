```python
# get dict with model names, area, and landfrac variable data for all datasets
datasets = []
for i in range(6):
    area_data = xr.open_dataset(area_paths[i])
    land_data = xr.open_dataset(land_paths[i])
    
    datasets.append({
        "model": model_names[i],
        "area": area_data,
        "landfrac": land_data
    })
```

```python
time_slice = slice("2085-01-01", "2099-12-30") # define timeslice (future)

# get temp data dicts for each future scenario
tas_ssp5_dict = {}
tas_ssp2_dict = {}
tas_g6sulfur_dict = {}

# --- SSP5 ---
for model, path in zip(model_names, ssp5_temp_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        tas_ssp5 = ds['tas'].sel(time=time_slice) - 273.15  # slice + convert to °C
        tas_ssp5_dict[model] = tas_ssp5
    except Exception as e:
        print(f"Error processing {model} (SSP5): {e}")

# --- SSP2 ---
for model, path in zip(model_names, ssp2_temp_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        tas_ssp2 = ds['tas'].sel(time=time_slice) - 273.15
        tas_ssp2_dict[model] = tas_ssp2
    except Exception as e:
        print(f"Error processing {model} (SSP2): {e}")

# --- G6sulfur ---
for model, path in zip(model_names, g6sulfur_temp_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        tas_g6sulfur = ds['tas'].sel(time=time_slice) - 273.15
        tas_g6sulfur_dict[model] = tas_g6sulfur
    except Exception as e:
        print(f"Error processing {model} (G6sulfur): {e}")
```

```python
# get pr data dicts for each future scenario
pr_ssp5_dict = {}
pr_ssp2_dict = {}
pr_g6sulfur_dict = {}

# --- SSP5 ---
for model, path in zip(model_names, pr_ssp5_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        pr_ssp5 = ds['pr'].sel(time=time_slice) * (60*60*24*360)
        pr_ssp5_dict[model] = pr_ssp5
    except Exception as e:
        print(f"Error processing {model} (SSP5): {e}")

# --- SSP2 ---
for model, path in zip(model_names, pr_ssp2_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        pr_ssp2 = ds['pr'].sel(time=time_slice) * (60*60*24*360)
        pr_ssp2_dict[model] = pr_ssp2
    except Exception as e:
        print(f"Error processing {model} (SSP2): {e}")

# --- G6sulfur ---
for model, path in zip(model_names, pr_g6_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        pr_g6 = ds['pr'].sel(time=time_slice) * (60*60*24*360)
        pr_g6sulfur_dict[model] = pr_g6
    except Exception as e:
        print(f"Error processing {model} (G6sulfur): {e}")
```

```python
time_slice_h = slice("1985-01-01", "2014-12-30") # define historical timeslice
# get tas and pr dicts for historical data
tas_hist_dict = {}
pr_hist_dict = {}

# pr
for model, path in zip(model_names, pr_h_paths):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        pr_hist = ds['pr'].sel(time=time_slice_h) * (60*60*24*360)
        pr_hist_dict[model] = pr_hist
    except Exception as e:
        print(f"Error processing {model} (Historical): {e}")

# tas
for model, path in zip(model_names, temp_paths_h):
    try:
        ds = xr.open_dataset(path, decode_times=True, use_cftime=True)
        tas_hist = ds['tas'].sel(time=time_slice_h) - 273.15
        tas_hist_dict[model] = tas_hist
    except Exception as e:
        print(f"Error processing {model} (Historical): {e}")
```

```python
land_thresh = {
    'CESM2-WACCM': 50,
    'CNRM-ESM2-1': 45,
    'IPSL-CM6A-LR': 30,
    'MPI-ESM1-2-LR': 35,
    'MPI-ESM1-2-HR': 50,
    'UKESM1-0-LL': 40
}

decadal_means = {
    'ssp5': {},
    'ssp2': {},
    'g6sulfur': {},
    'historical': {}
}

decadal_means_pr = {
    'ssp5': {},
    'ssp2': {},
    'g6sulfur': {},
    'historical': {}
}

for i, data in enumerate(datasets):
#get all the data from each type
    model = data['model']
    area_ds = data['area']
    land_ds = data['landfrac']
#specify variable, loop thru thresholds based on model and apply
    area_grid = area_ds['areacella']
    land_grid = land_ds['sftlf']
    threshold = land_thresh.get(model)
    land_mask = land_grid >= 0 #****IMPORTANT
    area_land = area_grid.where(land_mask)

    #USE THESE VARIABLES FROM NOW ON for avgs
    data["area_land"] = area_land

# tas
        # SSP5
    tas_ssp5 = tas_ssp5_dict[model].interp_like(area_land)
    decadal_ssp5 = (
        tas_ssp5.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means['ssp5'][model] = float(decadal_ssp5.values)

    # SSP2
    tas_ssp2 = tas_ssp2_dict[model].interp_like(area_land)
    decadal_ssp2 = (
        tas_ssp2.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means['ssp2'][model] = float(decadal_ssp2.values)

    # G6sulfur
    tas_g6 = tas_g6sulfur_dict[model].interp_like(area_land)
    decadal_g6 = (
        tas_g6.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means['g6sulfur'][model] = float(decadal_g6.values)

        # Historical
    tas_hist = tas_hist_dict[model].interp_like(area_land)
    decadal_hist = (
        tas_hist.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means['historical'][model] = float(decadal_hist.values)

# pr
        # SSP5
    pr_ssp5 = pr_ssp5_dict[model].interp_like(area_land)
    decadal_ssp5_pr = (
        pr_ssp5.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means_pr['ssp5'][model] = float(decadal_ssp5_pr.values)

        # SSP2
    pr_ssp2 = pr_ssp2_dict[model].interp_like(area_land)
    decadal_ssp2_pr = (
        pr_ssp2.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means_pr['ssp2'][model] = float(decadal_ssp2_pr.values)

        # G6sulfur
    pr_g6 = pr_g6sulfur_dict[model].interp_like(area_land)
    decadal_g6_pr = (
        pr_g6.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means_pr['g6sulfur'][model] = float(decadal_g6_pr.values)

            # Historical
    pr_hist = pr_hist_dict[model].interp_like(area_land)
    decadal_hist_pr = (
        pr_hist.weighted(area_land.fillna(0))
        .mean(dim=("lat", "lon"))
        .mean(dim="time")
    )
    decadal_means_pr['historical'][model] = float(decadal_hist_pr.values)

print("computations finished :)")
```

```python
#calculate temp anomalies for each scenario

ssp5_anomaly= {
    model: decadal_means['ssp5'][model] - decadal_means['historical'][model]
    for model in decadal_means['ssp5'] }

ssp2_anomaly= {
    model: decadal_means['ssp2'][model] - decadal_means['historical'][model]
    for model in decadal_means['ssp2'] }

g6_anomaly= {
    model: decadal_means['g6sulfur'][model] - decadal_means['historical'][model]
    for model in decadal_means['g6sulfur'] }
```

```python
#calculate pr anomalies for each scenario

ssp5_anomaly_pr= {
    model: ((decadal_means_pr['ssp5'][model] - decadal_means_pr['historical'][model])/decadal_means_pr['historical'][model]) * 100
    for model in decadal_means_pr['ssp5'] }

ssp2_anomaly_pr= {
    model: ((decadal_means_pr['ssp2'][model] - decadal_means_pr['historical'][model])/decadal_means_pr['historical'][model]) * 100
    for model in decadal_means_pr['ssp2'] }

g6_anomaly_pr= {
    model: ((decadal_means_pr['g6sulfur'][model] - decadal_means_pr['historical'][model])/decadal_means_pr['historical'][model]) * 100
    for model in decadal_means_pr['g6sulfur'] }
```

```python
# create figure

plt.rcParams['font.family'] = 'Times New Roman'
markers = ['o', 's', '^', 'D', 'v', 'P', '*', 'X']
models = list(ssp5_anomaly_pr.keys())
model_markers = {model: markers[i % len(markers)] for i, model in enumerate(models)}

scenario_colors = {
    "ssp5": "red",
    "ssp2": "limegreen",
    "g6sulfur": "royalblue"
}

def plot_scenario(x_dict, y_dict, color):
    for model in y_dict:
        plt.scatter(
            x_dict[model],
            y_dict[model],
            color=color,               
            marker=model_markers[model], 
            s=80,                      
            edgecolor='black',     
            linewidth=0.6,
            zorder=5
        )

fig, ax = plt.subplots(figsize=(7,5))

plot_scenario(ssp5_anomaly, ssp5_anomaly_pr, scenario_colors["ssp5"])
plot_scenario(ssp2_anomaly, ssp2_anomaly_pr, scenario_colors["ssp2"])
plot_scenario(g6_anomaly, g6_anomaly_pr, scenario_colors["g6sulfur"])

# axes through zero
ax.axhline(0, color='steelblue', linewidth=1, zorder=1)
ax.axvline(0, color='steelblue', linewidth=1,zorder=1)

ax.set_xlabel("Temperature Anomaly (°C)")
ax.set_ylabel("Precipitation Anomaly (%)")
ax.grid(True, alpha=0.3, color='steelblue')
ax.set_xticks([0, 2, 4, 6, 8])

scenario_handles = [
    Line2D([0], [0], marker='o', color='w', label='SSP5-8.5',
           markerfacecolor='red', markersize=8, markeredgecolor='black', markeredgewidth=0.6),
    Line2D([0], [0], marker='o', color='w', label='SSP2-4.5',
           markerfacecolor='limegreen', markersize=8, markeredgecolor='black', markeredgewidth=0.6),
    Line2D([0], [0], marker='o', color='w', label='G6sulfur',
           markerfacecolor='royalblue', markersize=8, markeredgecolor='black', markeredgewidth=0.6)
]

model_handles = [
    Line2D([0], [0], marker=model_markers[m], color='w', label=m,
           markerfacecolor='black', markersize=8,markeredgewidth=0.6)
    for m in models
]

scenario_labels = [h.get_label() for h in scenario_handles]
model_labels = [h.get_label() for h in model_handles]

fig.legend(
    handles=scenario_handles,
    labels=scenario_labels,
    title="Emissions Scenarios",
    loc='upper left',
    bbox_to_anchor=(1.02, 0.7),
    fontsize=9,
    title_fontsize=10,
    ncol=1,
    frameon=True
)

fig.legend(
    handles=model_handles,
    labels=model_labels,
    title="Climate Models",
    loc='upper left',
    bbox_to_anchor=(1.02, 0.5),
    fontsize=9,
    title_fontsize=10,
    ncol=1,
    frameon=True
)

plt.rcParams["font.size"] = 12
plt.rcParams["axes.labelsize"] = 12    
plt.rcParams["axes.titlesize"] = 15  
plt.rcParams["legend.fontsize"] = 11    
plt.rcParams["legend.title_fontsize"] = 12 


plt.tight_layout()
# save figure
plt.savefig('insert image path here', dpi=600, bbox_inches='tight')
plt.show()
```
